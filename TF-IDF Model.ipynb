### Import Statements

```python
#import statements
import pandas as pd

# Import TfidfVectorizer from sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

### Convert CSV file to Dataframe

```python
songs_df = pd.read_csv ('data.csv')
print(songs_df.head())
```

#### Remove the duplicate songs from the dataset

```python
songs_df = songs_df.drop_duplicates(subset=['id'])
songs_df = songs_df.reset_index(drop=True)
```

### Analyse the Data

```python
songs = songs_df.reset_index(drop=True)

# Replace \n present in the text
songs['text'] = songs['lyrics'].str.replace(r'\n', '')

# Initialize tfidf vectorizer
tfidf = TfidfVectorizer(analyzer='word', stop_words='english')

# Fit and transform 
tfidf_matrix = tfidf.fit_transform(songs['text'])
```

```python
tfidf_matrix
```

### Building Model using Cosine Similarity

```python
cosine_similarities = cosine_similarity(tfidf_matrix)
```

```python
len(cosine_similarities)
```

```python
similarities = {}

for i in range(len(cosine_similarities)):
    # Get the indexes of the songs after Sorting all the elements using cosine similarities 
    similar_indices = cosine_similarities[i].argsort()[:-7:-1] 
    #Store the 5 most similar songs
    similarities[songs_df['title'].iloc[i]] = [(cosine_similarities[i][x], songs_df['title'][x], songs_df['artist'][x]) for x in similar_indices][1:]
```

```python
similarities
```
